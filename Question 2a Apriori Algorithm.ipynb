## Apriori Algorithm

```python
def candidate_prune(Ck, Fk):
    T = set({})
    for x in Ck:
        c=0
        for y in x:
            if x-{y} not in Fk:
                T.add(x)
                break
    return Ck-T
```

```python
def candidate_gen(Fk):
    T=set({})
    for x in Fk:
        k=len(x)
        for y in Fk:
            if(len(x&y)==k-1):
                T.add(frozenset(x|y))
    return T
```
